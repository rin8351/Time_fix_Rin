```python
import time as t1
import datetime as t2
import ast
from kivy.uix.button import Button
from kivy.uix.label import Label
from kivy.uix.boxlayout import BoxLayout
from kivy.uix.scrollview import ScrollView
from kivy.core.window import Window
from kivy.uix.textinput import TextInput
import os
from kivy.base import runTouchApp
from kivy.network.urlrequest import UrlRequest
import urllib
a='text' # переменная для функции

class Fix_time(BoxLayout):
    def __init__(self, **kwargs):
        super(Fix_time, self).__init__(**kwargs)

        if os.path.isfile('fix.txt')==True:
            pass
        else:
            with open('fix.txt','w') as f:
                f.write('')
                
            self.writ('fix.txt','0','a')
            self.writ('fix.txt','Нефикс','a')
            dlya_zapisi=str({'Нефикс': '00:00:00', 'Еда': '00:00:00', 'Сон': '00:00:00','Быт': '00:00:00','Отдых': '00:00:00','Хобби': '00:00:00','Работа': '00:00:00'}) #Список занятий, котрый можно дополнять
            # Нефикс- сокращенно от нефиксированное время
            self.writ('fix.txt', dlya_zapisi,'a')

        self.spisok=[] 
        for i in (ast.literal_eval((self.writ('fix.txt',a,'r'))[-1])):
            self.spisok.append(i)

        self.size_hint=(1, None)
        self.size=(Window.width, Window.height)
        self.orientation="horizontal"
        
        self.pole1 =BoxLayout(orientation='vertical')
        self.pole1.bind(minimum_height = self.setter('height'))
        self.pole1_1= BoxLayout(orientation='vertical')
        self.pole1_1.bind(minimum_height = self.setter('height'))
        
        self.pole2= BoxLayout(orientation='vertical')
        self.pole2.bind(minimum_height = self.setter('height'))
        self.pole2_2= BoxLayout(orientation='vertical')
        self.pole2_2.bind(minimum_height = self.setter('height'))
        
        self.pole3= BoxLayout(orientation='vertical')
        self.pole3.bind(minimum_height = self.setter('height'))
        
        self.btn_zan()
        self.add_widget(self.pole1)  
        
        self.table()
        self.add_widget(self.pole2)
        


        self.vbr_string=(((self.writ('fix.txt',a,'r'))[1]).strip())
        self.vbr=Label(text='Выбранное:\n '+ self.vbr_string)
        self.pole3.add_widget(self.vbr)
        
        self.btn1=Button(text="Сколько\n длится")
        self.btn1.bind(on_press=self.skolko_dlitsya)
        self.pole3.add_widget(self.btn1)
        
        self.lab1=Label(text='0')
        self.pole3.add_widget(self.lab1)
        
        self.btn3=Button(text="Новый день")
        self.btn3.bind(on_press=self.Noviy)
        self.pole3.add_widget(self.btn3)
        
        self.lab2=Label(text='')
        self.pole3.add_widget(self.lab2)
        
        self.btn4=Button(text="Отправка\n архива")
        self.btn4.bind(on_press=self.send_archive)
        self.pole3.add_widget(self.btn4)
        
        self.lab3=Label(text='Поле для ввода\nнового занятия')
        self.pole3.add_widget(self.lab3)
        
        self.txtinp=TextInput(text='')
        self.pole3.add_widget(self.txtinp)
        
        self.btn5=Button(text="Добавить новое")
        self.btn5.bind(on_press=self.dob)
        self.pole3.add_widget(self.btn5)
        
        self.lab3=Label(text='Поле для удаления\n занятия.')
        self.pole3.add_widget(self.lab3)
        
        self.lab4=Label(text='')
        self.pole3.add_widget(self.lab4)
        
        self.btn6=Button(text="Удалить выбранное\nзанятие")
        self.btn6.bind(on_press=self.udal)
        self.pole3.add_widget(self.btn6)
        
        self.lab5=Label(text='')
        self.pole3.add_widget(self.lab5)
                      
        
        self.add_widget(self.pole3)

        self.zan=ast.literal_eval((self.writ('fix.txt',a,'r'))[-1])
  
    def table(self):
        self.pole2.remove_widget(self.pole2_2)
        self.pole2_2= BoxLayout(orientation='vertical')
        self.pole2_2.bind(minimum_height = self.setter('height'))
        self.sp=ast.literal_eval((self.writ('fix.txt',a,'r'))[-1])
        for i in self.sp:
            self.lbl_tb=Label(text=self.sp.get(i))
            self.lbl_tb.bind(size=self.lbl_tb.setter('text_size'))
            self.pole2_2.add_widget(self.lbl_tb)
        self.pole2.add_widget(self.pole2_2)
        return
        
        #halign="left", valign="middle"


    def btn_zan(self):
        self.pole1.remove_widget(self.pole1_1)
        self.pole1_1= BoxLayout(orientation='vertical')
        self.pole1_1.bind(minimum_height = self.setter('height'))
        for i in self.spisok:
            j=Button(text=str(i), on_press=self.Press_button,size_hint_x= 1,size_hint_y= None,size=(160,150))
            self.pole1_1.add_widget(j)
        self.pole1.add_widget(self.pole1_1)

    def Press_button(self,instance): 
        self.vbr_zan=str(instance.text)
        self.lab2.text=''
        if self.vbr_zan=='Новый день':
            self.vbr_zan='Нефикс'
        self.end=int(((self.writ('fix.txt',a,'r'))[0]).strip())
        self.start=self.time_now()
        if self.start>=self.end:
            self.vremya=self.start-self.end
        else:
            self.vremya=(86399-self.end)+self.start

        self.vibor= ((self.writ('fix.txt',a,'r'))[1]).strip()
        format ='%H:%M:%S'
        zan_time= t1.strptime(self.zan[self.vibor], format)
        zan_sec= (zan_time.tm_hour*3600)+(zan_time.tm_min*60)+(zan_time.tm_sec)
        self.zan[self.vibor]=str(t2.timedelta(seconds = (self.vremya + zan_sec))) 

        self.writ('fix.txt',str(self.start),'w')
        self.writ('fix.txt',str(self.start),'a')
        self.writ('fix.txt',str(self.vbr_zan),'a')
        self.writ('fix.txt',str(self.zan),'a')   

        self.table()
        self.vbr_string=(((self.writ('fix.txt',a,'r'))[1]).strip())
        self.vbr.text='Выбранное:\n '+ self.vbr_string
        
        self.lab4.text=self.vbr_string

    def skolko_dlitsya(self,event):
        self.end=int(((self.writ('fix.txt',a,'r'))[0]).strip())
        self.start=self.time_now()
        if self.start>self.end:
            self.vremya=self.start-self.end
        else:
            self.vremya=(86399-self.end)+self.start
        string_time=str(t2.timedelta(seconds = (self.vremya)))
        self.lab1.text=string_time
        self.lab2.text=''


    def Noviy(self,instance):
        try:
            self.Press_button(instance)
            self.lab2.text='Отсчет\n нового дня'        
            self.vbr_string=(((self.writ('fix.txt',a,'r'))[1]).strip())
            self.vbr.text='Выбранное:\n '+ self.vbr_string

            today = t2.date.today()
            str_td=str(today.year)+'.'+str(today.month)+'.'+str(today.day)
            string= str_td + ' '+str(self.zan)+'\n'

            
            if os.path.isfile('archive.txt')==True:
                pass
            else:
 
                with open('archive.txt','w') as f:
                    f.write('')

            self.writ('archive.txt',string,'a')

            self.vrem=self.zan.copy()
            for i in self.vrem:
                self.vrem[i]= '00:00:00'
            self.start=self.time_now()

            self.writ('fix.txt',str(self.start),'w')
            self.writ('fix.txt',str(self.start),'a')
            self.writ('fix.txt',str(self.vbr_zan),'a')        
            self.writ('fix.txt',str(self.vrem),'a')
            self.zan=ast.literal_eval((self.writ('fix.txt',a,'r'))[-1])
            self.lab1.text=''

            self.table() 

        except Exception as e:
            self.vbr.text='err: '+str(e)
            
    def send_archive(self,event):
        url	= ''# здесь указан номер сервера своего
        with open('archive.txt', 'r') as f:
            data = str(f.read())
        headers = {'Content-type': 'application/x-www-form-urlencoded','Accept': 'text/plain'}
        req = UrlRequest(url, req_body=urllib.parse.urlencode({'file': data}), req_headers=headers)

    def time_now(self):
        hours=int(t1.strftime("%H", t1.localtime()))
        minut=int(t1.strftime("%M", t1.localtime()))
        sec=int(t1.strftime("%S", t1.localtime()))
        ti=(hours*3600)+(minut*60)+sec
        return ti   


    def writ(self,txt,strr,bukv):
        result =''
        with open(txt,bukv) as file:
            if bukv=='a':
                file.write(strr +'\n')
            if bukv=='w':
                file.write('')
            if bukv=='r':
                result = file.read().splitlines()
        return result
        
    def dob(self,event):

        self.spisok.append(self.txtinp.text)
        
        self.btn_zan()
    
        self.zan[self.txtinp.text]='00:00:00'
        self.writ('fix.txt',str(self.zan),'a')  
        self.table()
        
    def udal(self,event):
        if self.vbr_string=='Нефикс':
            self.lab5.text='Это занятие удалять\nнельзя,выберите\n другое'
        else:
            self.lab5.text=''
            for i in self.spisok:
                if i==self.vbr_string:
                    self.spisok.remove(i)
            
            self.btn_zan()
            
            self.vbr_string='Нефикс'
            self.vbr.text='Выбранное:\n '+ self.vbr_string
            format ='%H:%M:%S'
            zan_time2= t1.strptime(self.zan['Нефикс'], format)
            zan_sec2= (zan_time2.tm_hour*3600)+(zan_time2.tm_min*60)+(zan_time2.tm_sec)
            print('zan_sec2',zan_sec2)
            
            self.end=int(((self.writ('fix.txt',a,'r'))[0]).strip())
            self.start=self.time_now()
            if self.start>=self.end:
                self.vremya=self.start-self.end
            else:
                self.vremya=(86399-self.end)+self.start
            print('vremya',self.vremya)
            
            self.zan['Нефикс']=str(t2.timedelta(seconds = (zan_sec2+self.vremya))) 
            self.zan.pop(self.vbr_zan)
            self.writ('fix.txt',str(self.start),'w')
            self.writ('fix.txt',str(self.start),'a')
            self.writ('fix.txt','Нефикс','a')
            self.writ('fix.txt',str(self.zan),'a') 
            self.lab5.text=''
            self.lab4.text=''

            self.table()
            
root = ScrollView()
root.add_widget(Fix_time())

runTouchApp(root)
```
